## 2. Statistik Deskriptif

Statistik deskriptif ada 2, untuk melihat pemusatan data dan penyebaran data.

* **Pemusatan data**: Mean, median, modus
* **Penyebaran data**: Varian, standar deviasi

### a. Pemusatan Data

#### 1) Mean

```python
import matplotlib.pyplot as plt
print(f"Data: \n{games_data['Rating']}")

mean_value=games_data['Rating'].mean()

print(f"\nMean: {mean_value: .2f}")
```

Nilai mean kemudian divisualisasikan dalam format grafik sebagai berikut.

```python
plt.hist(games_data['Rating'], bins=30, edgecolor='black', alpha=0.5, color='blue')

plt.axvline(x=mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')

plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Mean')

plt.legend()

plt.show()
```

#### 2) Median

```python
import matplotlib.pyplot as plt
import pandas as pd
print(f"Data: \n{games_data['Rating']}")

sorted_data = sorted(games_data['Rating'])
print(f"\nData berurut: {sorted_data}")

n = len(sorted_data)
print(f"\nJumlah total data: {n}")

median_value=games_data['Rating'].median()

print(f"\nMedian: {median_value}")
```

Nilai median kemudian divisualisasikan dalam format grafik sebagai berikut.

```python
plt.hist(games_data['Rating'], bins=30, edgecolor='black', alpha=0.5, color='blue')

plt.axvline(x=median_value, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value:.2f}')

plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Median')

plt.legend()

plt.show()
```

#### 3) Modus

```python
import matplotlib.pyplot as plt
from collections import Counter

print(f"Data: \n{games_data['Rating']}")

counter = Counter(games_data['Rating'])
mode_data = dict(counter)
max_counter = max(list(counter.values()))
print(f"\nMax Counter: {max_counter}")

mode_value = []
for k, v in mode_data.items():
    print(f"{k} muncul {v} kali")
    if v == max(list(counter.values())):
        mode_value.append(k)
    
print(f"\nModus: {mode_value}")
```

Nilai Modus kemudian divisualisasikan dalam format grafik sebagai berikut.

```python
plt.hist(games_data['Rating'], bins=30, edgecolor='black', alpha=0.5, color='green')

plt.axvline(x=mode_value[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus = {mode_value[0]:.2f}')

plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Modes')

plt.legend()

plt.show()
```

Berdasarkan perhitungan yang telah dilakukan, didapatin nilai mean, median, dan modus berturut-turut adalah 3.72, 3.80, dan 4.10 sehingga dapat disimpulkan bahwa nilai mean dan median tidak dipengaruhi oleh outliers sedangkan untuk modus juga cukup sesuai dengan kedua nilai sebelumnya dikarenakan memiliki nilai yang mendekati dan distribusi datanya dapat disebut normal.

### b.Penyebaran Data

#### 1) Range

```python
import matplotlib.pyplot as plt

print(f"Data: \n{games_data['Rating']}")
lowest=games_data['Rating'].min()
highest=games_data['Rating'].max()

print(f"\nlowest: {lowest}")
print(f"\nhighest: {highest}")
```

```python
plt.hist(games_data['Rating'], bins=30, edgecolor='black', alpha=0.5, color='green')

plt.axvline(x=lowest, color='red', linestyle='dashed', linewidth=1, label=f'min data = {lowest:.2f}')
plt.axvline(x=highest, color='red', linestyle='dashed', linewidth=1, label=f'max data = {highest:.2f}')

plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Range')

plt.legend()

plt.show()
```

#### 2) Standar Deviasi

```python
import matplotlib.pyplot as plt

print(f"Data: \n{games_data['Rating']}")
standard_deviation = games_data['Rating'].std()

print(f"\nStandar Deviasi= {standard_deviation: .2f}")
```

```python
plt.hist(games_data['Rating'], bins=30, edgecolor='black', alpha=0.5, color='green')

plt.axvline(x=mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value: .2f}')
plt.axvline(x=mean_value - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean_value-standard_deviation:.2f}')
plt.axvline(x=mean_value + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean_value+standard_deviation:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Standard Deviation = {standard_deviation:.2f}')

plt.legend()

plt.show()
```

Didapati bahwa kebanyakan game memiliki rating antara 3.19 - 4.25 yang menandakan bahwa kemungkinan besar ada lebih banyak game yang dimainkan dengan jangkauan rating ini.

## 3. Analisis Hubungan Atribut

### a) Release Date

Pada bagian ini dilakukan pengolahan data lanjutan yang mengabaikan nilai "releases on TBD" karena game yang belum dirilis tidak memiliki rating dan akan mengganggu proses analisis apabila tidak diabaikan.

```python
import matplotlib.pyplot as plt
import pandas as pd

dropTBD = games_data.drop(games_data[games_data['Release Date'] == 'releases on TBD'].index)
sortPlay=dropTBD.sort_values(by='Plays',ascending=True)
playby=sortPlay.tail(6)

plt.barh(playby['Release Date'],playby['Plays'])
plt.ylabel("Realese Date")
plt.xlabel("Play Times")
plt.show()
```

Dapat dilihat bahwa tanggal perilisan suatu game sama sekali tidak berpengaruh terhadap seberapa banyak game tersebut dimainkan. Game dengan jumlah dimainkan terbanyak dirilis pada tahun 2010, diikuti 1994, 2005, 2010 lagi, 1995, dst. Sehingga pola dari hubungan ini tidak dapat ditemukan.

### b) Rating

Karena sebelumnya sempat ditemukan nilai "null" di dalam atribut ini sehingga akan dilakukan pengolahan data lanjutan sebelum analisis.

```python
import matplotlib.pyplot as plt
import pandas as pd

games_data.dropna(subset=['Rating'],inplace=True)
playby=sortPlay.tail(7)

plt.bar(playby['Plays'],playby['Rating'],width=0.4)
plt.ylabel("Rating")
plt.xlabel("Play Times")
plt.show()
```

Berdasarkan perhitungan sebelumnya di bagian "Statistik Deskriptif" kebanyakan game memiliki jangkauan rating antara 3.19 - 4.25, dan dapat dilihat bahwa memang jika dihubungkan dengan jumalh dimainkan akan terlihat cukup sesuai, namun, pola yang dihasilkan tetap tidak bisa ditemukan karena seperti contoh di atas terlihat bahwa game dengan rating di atas 4 dengan jumlah dimainkan sebanyak 984 kali bisa dikalahkan oleh game yang bahkan hanya melewati rating 3.

### c) Genre

```python
import matplotlib.pyplot as plt
import pandas as pd

playby=sortPlay.tail(7)


plt.barh(playby['Genres'],playby['Plays'])
plt.xlabel("Play Times")
plt.show()
```

Karena genre pada setiap game tidak dapat dipisahkan per-elemennya, sehingga dapat dilihat bahwa game yang paling banyak dimainkan kebanyakan memiliki genre antara "Adventure" dan "RPG" karena muncul hingga 3 kali di atas.

### d) Playing

```python
import matplotlib.pyplot as plt
import pandas as pd

playby=sortPlay.tail(7)


plt.barh(playby['Playing'],playby['Plays'])
plt.ylabel("Active Player(s)")
plt.xlabel("Play Times")
plt.show()
```

Sementara itu, jumlah pemain aktif di sini sangat membuktikan bahwa tidak ada hubungan sama sekali antara pemain aktif dengan jumlah suatu game dimainkan. Bahkan, game yang memiliki 1 pemain aktif sudah dimainkan lebih banyak dari pada game dengan pemain aktif 126.
